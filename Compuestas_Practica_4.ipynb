```python
import numpy as np
import pandas as pd

#Define la función a calcular
def f1(x):
  return np.exp(x**2)
def f2(x):
  return np.sin(x**2)
def f3(x):
  return np.exp(np.cos(x))
def f4(x):
  return np.exp(2*x)*(np.sin(3*x))


#Valores reales de las integrales
vr1 = 1.462651745907
vr2 = 0.8948314678778
vr3 = 3.977463256154
vr4 = -14.2139771296096

#Trapecio Multiple

def TC(a,b,n,f):
  h = (b-a)/n
  x = np.zeros([n])

  for i in range(0,n):
    x[i] = a + i*(h)

  sum = 0
  for i in range(1,n):
    sum = sum + f(x[i])

  I = (h/2)*(f(a) + 2*(sum) + f(b))
  return I

#Simpson Multiple

def S(a,b,n,f):
  h = (b-a)/n
  x = np.zeros([n+1])

  for i in range(0,n+1):
    x[i] = a + i*(h)

  sum1 = 0
  sum2 = 0

  for i in range(1,n,2):
    sum1 = sum1 + f(x[i])
  for i in range(2,n,2):
    sum2 = sum2 + f(x[i])
  I = (h/3)*(f(a)+ 4*sum1 + 2*sum2 + f(b))
  return I

#Errores

def ER(vr,va):
  er = abs(vr-va)/abs(vr)
  return er

def EA(vr,va):
  ea = abs(vr-va)
  return ea

print(' --- Problema 1 ---')
print(f'Valor real {vr1}')
df = pd.DataFrame()
df['Método'] = ['Trapecio n = 2', 'Trapecio n = 4','Trapecio n = 8','Trapecio n = 16', 'Simpson n = 2','Simpson n = 4','Simpson n = 8','Simpson n = 16' ]
df['Aproximación'] = [TC(0,1,2,f1), TC(0,1,4,f1), TC(0,1,8,f1),TC(0,1,16,f1), S(0,1,2,f1), S(0,1,4,f1),S(0,1,8,f1), S(0,1,16,f1)]
df["Error relativo"] = [ER(vr1,TC(0,1,2,f1)),ER(vr1,TC(0,1,4,f1)),ER(vr1,TC(0,1,8,f1)),ER(vr1,TC(0,1,16,f1)),ER(vr1,S(0,1,2,f1)),ER(vr1,S(0,1,4,f1)),ER(vr1,S(0,1,8,f1)),ER(vr1,S(0,1,16,f1))]
df["Error absoluto"] = [EA(vr1,TC(0,1,2,f1)),EA(vr1,TC(0,1,4,f1)),EA(vr1,TC(0,1,8,f1)),EA(vr1,TC(0,1,16,f1)),EA(vr1,S(0,1,2,f1)),EA(vr1,S(0,1,4,f1)),EA(vr1,S(0,1,8,f1)),EA(vr1,S(0,1,16,f1))]
print(df)
print(' ')
print(' --- Problema 2 ---')
print(f'Valor real {vr2}')
dl = pd.DataFrame()
dl['Método'] = ['Trapecio n = 2', 'Trapecio n = 4','Trapecio n = 8','Trapecio n = 16', 'Simpson n = 2','Simpson n = 4','Simpson n = 8','Simpson n = 16' ]
dl['Aproximación'] = [TC(0,np.sqrt(np.pi),2,f2), TC(0,np.sqrt(np.pi),4,f2), TC(0,np.sqrt(np.pi),8,f2),TC(0,np.sqrt(np.pi),16,f2), S(0,np.sqrt(np.pi),2,f2), S(0,np.sqrt(np.pi),4,f2),S(0,np.sqrt(np.pi),8,f2), S(0,np.sqrt(np.pi),16,f2)]
dl["Error relativo"] = [ER(vr2,TC(0,np.sqrt(np.pi),2,f2)),ER(vr2,TC(0,np.sqrt(np.pi),4,f2)),ER(vr2,TC(0,np.sqrt(np.pi),8,f2)),ER(vr2,TC(0,np.sqrt(np.pi),16,f2)),ER(vr2,S(0,np.sqrt(np.pi),2,f2)),ER(vr2,S(0,np.sqrt(np.pi),4,f2)),ER(vr2,S(0,np.sqrt(np.pi),8,f2)),ER(vr2,S(0,np.sqrt(np.pi),16,f2))]
dl["Error absoluto"] = [EA(vr2,TC(0,np.sqrt(np.pi),2,f2)),EA(vr2,TC(0,np.sqrt(np.pi),4,f2)),EA(vr2,TC(0,np.sqrt(np.pi),8,f2)),EA(vr2,TC(0,np.sqrt(np.pi),16,f2)),EA(vr2,S(0,np.sqrt(np.pi),2,f2)),EA(vr2,S(0,np.sqrt(np.pi),4,f2)),EA(vr2,S(0,np.sqrt(np.pi),8,f2)),EA(vr2,S(0,np.sqrt(np.pi),16,f2))]
print(dl)
print(' ')
print(' --- Problema 3 ---')
print(f'Valor real {vr3}')
ds = pd.DataFrame()
ds['Método'] = ['Trapecio n = 2', 'Trapecio n = 4','Trapecio n = 8','Trapecio n = 16', 'Simpson n = 2','Simpson n = 4','Simpson n = 8','Simpson n = 16' ]
ds['Aproximación'] = [TC(0,np.pi,2,f3), TC(0,np.pi,4,f3), TC(0,np.pi,8,f3),TC(0,np.pi,16,f3), S(0,np.pi,2,f3), S(0,np.pi,4,f3),S(0,np.pi,8,f3), S(0,np.pi,16,f3)]
ds["Error relativo"] = [ER(vr3,TC(0,np.pi,2,f3)),ER(vr3,TC(0,np.pi,4,f3)),ER(vr3,TC(0,np.pi,8,f3)),ER(vr3,TC(0,np.pi,16,f3)),ER(vr3,S(0,np.pi,2,f3)),ER(vr3,S(0,np.pi,4,f3)),ER(vr3,S(0,np.pi,8,f3)),ER(vr3,S(0,np.pi,16,f3))]
ds["Error absoluto"] = [EA(vr3,TC(0,np.pi,2,f3)),EA(vr3,TC(0,np.pi,4,f3)),EA(vr3,TC(0,np.pi,8,f3)),EA(vr3,TC(0,np.pi,16,f3)),EA(vr3,S(0,np.pi,2,f3)),EA(vr3,S(0,np.pi,4,f3)),EA(vr3,S(0,np.pi,8,f3)),EA(vr3,S(0,np.pi,16,f3))]
print(ds)
print(' ')
print(' --- Problema 4 ---')
print(f'Valor real {vr4}')
dm = pd.DataFrame()
dm['Método'] = ['Trapecio n = 2', 'Trapecio n = 4','Trapecio n = 8','Trapecio n = 16', 'Simpson n = 2','Simpson n = 4','Simpson n = 8','Simpson n = 16' ]
dm['Aproximación'] = [TC(0,2,2,f4), TC(0,2,4,f4), TC(0,2,8,f4),TC(0,2,16,f4), S(0,2,2,f4), S(0,2,4,f4),S(0,2,8,f4), S(0,2,16,f4)]
dm["Error relativo"] = [ER(vr4,TC(0,2,2,f4)),ER(vr4,TC(0,2,4,f4)),ER(vr4,TC(0,2,8,f4)),ER(vr4,TC(0,2,16,f4)),ER(vr4,S(0,2,2,f4)),ER(vr4,S(0,2,4,f4)),ER(vr4,S(0,2,8,f4)),ER(vr4,S(0,2,16,f4))]
dm["Error absoluto"] = [EA(vr4,TC(0,2,2,f4)),EA(vr4,TC(0,2,4,f4)),EA(vr4,TC(0,2,8,f4)),EA(vr4,TC(0,2,16,f4)),EA(vr4,S(0,2,2,f4)),EA(vr4,S(0,2,4,f4)),EA(vr4,S(0,2,8,f4)),EA(vr4,S(0,2,16,f4))]
print(dm)



```
