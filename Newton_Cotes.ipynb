```python
import numpy as np

#Define la funci√≥n a calcular
def f1(x):
  return x**2*(np.log(x))
def f2(x):
  return x**3 + 2*x**2
def f3(x):
  return (np.exp(x)-1)/(np.sin(x))
def f4(x):
  return x*np.cos(x**2)

#Valores reales
w = 22.574279703891
x = 1.3333333333333
y = 2.4228504234839
z = -0.2151506084997



def Cerrado3(a,b,n,f):
  h = (b-a)/n
  x = np.zeros([n+1])

  for i in range(0,n+1):
    x[i] = a + i*(h)
    #print(f'x_{i} =', round(x[i],6))

  I = ((3*h)/8)*(f(x[0])+3*f(x[1])+3*f(x[2])+f(x[3]))

  return I

def Cerrado4(a,b,n,f):
  h = (b-a)/n
  x = np.zeros([n+1])
  for i in range(0,n+1):
    x[i] = a + i*(h)
    #print(f'x_{i} =', round(x[i],6))

  I = ((2*h)/45)*(7*f(x[0])+32*f(x[1])+12*f(x[2])+32*f(x[3])+7*f(x[4]))
  return I

def Abierto2(a,b,n,f):
  h = (b-a)/(n+2)
  x = np.zeros([n+3])
  for i in range(0,n+3):
    x[i] = a + i*(h)
    #print(f'x_{i-1} =', round(x[i],6))

  I = ((4*h)/3)*(2*f(x[1])-f(x[2])+2*f(x[3]))
  return I

def Abierto3(a,b,n,f):
  h = (b-a)/(n+2)
  x = np.zeros([n+3])
  for i in range(0,n+3):
    x[i] = a + i*(h)
    #print(f'x_{i-1} =', round(x[i],6))

  I = ((5*h)/24)*(11*f(x[1])+f(x[2])+f(x[3])+11*f(x[4]))
  return I

def Er(vr,va):
  e = abs(vr-va)
  return e
def Ea(vr,va):
  e = abs((vr-va)/vr)
  return e
```
